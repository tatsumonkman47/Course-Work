```python
import numpy as np
```

```python
a = 2*np.eye(20,20) - np.eye(20,20,1) - np.eye(20,20,-1) 
a[0][1] = -2
a[-1][-2] = -2
aa = 2*np.eye(10,10) - np.eye(10,10,1) - np.eye(10,10,-1) 
```

```python
w,v = np.linalg.eig(a)
print(w)
print(v)
```

```python
print(np.linalg.inv(a))
```

```python
print(np.linalg.det(a))
```
